```python
#3.1 Introduction of my Scenario

#Yayasan Sejahtera (YS) merupakan sebuah badan bukan kerajaan (NGO). 
#YS memberi fokus kepada usaha pembasmian kemiskinan di kalangan golongan berpendapatan rendah di seluruh Malaysia.
#YS menyediakan bantuan kepada individu, isi rumah dan komuniti yang memerlukan, dengan matlamat untuk meningkatkan kualiti hidup
#dan membina kemampanan ekonomi dalam jangka yang panjang.
#Bantuan yang diberikan oleh YS dapat dikategorikan kepada dua jenis utama iaitu bantuan secara satu kali (one-off) dan bantuan mata pencarian (livelihood).
#Bantuan one-off disasarkan untuk menampung keperluan segera individu, isi rumah atau komuniti yang menghadapi krisis, contohnya menghadapi bencana alam seperti banjir. 
#Bantuan jenis ini diberikan serta merta atau dalam tempoh yang pendek sahaja bagi memastikan keperluan kritikal penerima dapat dipenuhi dengan segera.
#Manakala, bantuan mata pencarian pula ia menumpukan kepada pembangunan komuniti yang mampan, dengan memberi peluang kepada individu, isi rumah, atau kumpulan untuk
#berdikari dan membina kemampanan ekonomi secara kukuh dan stabil.
#Bantuan jenis ini biasanya akan dilaksanakan dalam jangka masa 6 hingga ke 36 bulan, bergantung kepada kepada keperluan projek dan kapasiti penerima.
#Antara bantuan yang diberikan adalah bantuan modal perniagaan, latihan asas keusahawanan dan bantuan peralatan perniagaan,
#Sebagai seorang pengurus pangkalan data di YS, adalah penting untuk mempunyai sistem pengurusan data yang tersusun dan efisien,
#bagi merekodkan maklumat penerima bantuan, sama ada individu, isi rumah atau komuniti. 
#Sistem ini membolehkan YS memantau bantuan yang diberikan, menilai keberkesanan program, dan memastikan bahawa sumber bantuan diagihkan secara adil
#dan tepat kepada meraka yang memerlukan.
#Melalui konsep pengaturcaraan berorientasikan objek (OOP) seperti kelas (class), fungsi (function), dan pewarisan (inheritance)
#kita dapat memodelkan data penerima bantuan secara sistematik, di mana penerima individu, isi rumah dan komuniti boleh
#dikategorikan dengan attribut yang sesuai, disimpan dan dipaparkan dengan mudah.
```

```python
#3.2 Class, function and inheritance

# Parent class untuk penerima bantuan individu/isi rumah
class PenerimaBantuanIndividu_IsiRumah:
    def __init__(self, nama, umur, lokasi, negeri, nama_program, jenis_bantuan):
        self.nama = nama
        self.umur = umur
        self.lokasi = lokasi
        self.negeri = negeri
        self.nama_program = nama_program
        self.jenis_bantuan = jenis_bantuan

    def display_info(self):
        print(f"Nama                       : {self.nama}")
        print(f"Umur                       : {self.umur} tahun")
        print(f"Lokasi                     : {self.lokasi}")
        print(f"Negeri                     : {self.negeri}")
        print(f"Nama Program               : {self.nama_program}")
        print(f"Jenis Bantuan              : {self.jenis_bantuan}")
        print("-" * 50)


# Child class untuk penerima bantuan komuniti/kumpulan
class PenerimaBantuanKomuniti(PenerimaBantuanIndividu_IsiRumah):
    def __init__(self, nama, lokasi, negeri, nama_program, jenis_bantuan,
                 jumlah_ahli_komuniti, jenis_komuniti):

        # Komuniti tiada umur â†’ letak None
        super().__init__(nama, umur=None, lokasi=lokasi, negeri=negeri,
                         nama_program=nama_program, jenis_bantuan=jenis_bantuan)

        self.jumlah_ahli_komuniti = jumlah_ahli_komuniti
        self.jenis_komuniti = jenis_komuniti

    def display_info(self):
        print(f"Komuniti                   : {self.nama}")
        print(f"Lokasi                     : {self.lokasi}")
        print(f"Negeri                     : {self.negeri}")
        print(f"Nama Program               : {self.nama_program}")
        print(f"Jenis Bantuan              : {self.jenis_bantuan}")
        print(f"Jumlah Ahli Komuniti       : {self.jumlah_ahli_komuniti}")
        print(f"Jenis Komuniti             : {self.jenis_komuniti}")
        print("-" * 50)
```
